```python
import pandas as pd
import numpy as np
import math
from tabulate import tabulate

def compute_metrics(df, current_sales, percentage_increase):
    # Calculate the new revenue target by increasing the current sales by the given percentage
    targets_need_to_achieve = current_sales * (1 + percentage_increase / 100)
    revenue_target = targets_need_to_achieve

    # Group by cardholder_id and calculate statistics for each customer
    grouped = df.groupby('cardholder_id').agg(
        Total_Transaction_Value=('transaction_amount', 'sum'),
        Total_Cashback_Value=('cashback_amount', 'sum')
    ).reset_index()

    # Calculate required metrics
    unique_customers = df['cardholder_id'].nunique()
    avg_order = grouped['Total_Transaction_Value'].mean()
    avg_cashback = grouped['Total_Cashback_Value'].mean()
    cashback_percentage = (avg_cashback / avg_order) * 100

    # Round off metrics for better presentation
    avg_order_rounded = round(avg_order, 2)
    avg_cashback_rounded = round(avg_cashback, 2)
    cashback_percentage_rounded = round(cashback_percentage, 2)

    # Calculate metrics outputs
    no_of_customers_to_target = revenue_target / (avg_order_rounded - avg_cashback_rounded)
    no_of_customers_to_target_rounded = math.ceil(no_of_customers_to_target)  # Always round up

    cashback_budget = no_of_customers_to_target_rounded * avg_cashback_rounded
    target_achieve = no_of_customers_to_target_rounded * avg_order_rounded
    profit = target_achieve - cashback_budget

    # Display the metrics
    print("Merchant Inputs:")
    print(f"Current Sales: {current_sales}")
    print(f"Targets Need to Achieve (Increased by {percentage_increase}%): {targets_need_to_achieve}")
    print(f"Revenue Target: {revenue_target}")
    print()
    
    print("Data:")
    print(f"No of Customers in Cluster: {unique_customers}")
    print(f"Avg Order: {avg_order_rounded}")
    print(f"Avg Cashback: {avg_cashback_rounded}")
    print(f"Cashback %: {cashback_percentage_rounded}%")
    print()
    
    print("Metrics Outputs:")
    print(f"No of Customers to Target: {no_of_customers_to_target_rounded} (Approx.)")
    print(f"Cashback Budget: {cashback_budget}")
    print(f"Target Achieve: {target_achieve}")
    print(f"Profit: {profit}")
    print()
    
    # Check if profit meets or exceeds the revenue target
    if profit >= revenue_target:
        print(f"Yes, we achieved the target successfully with approximately {no_of_customers_to_target_rounded} customers!")
        print('\n\n**********************************')
        print(f'Lets Try with Random {no_of_customers_to_target_rounded} Cardholders')
    else:
        print(f"Sorry, we cannot achieve the target with {no_of_customers_to_target_rounded} customers. Consider targeting more customers or increasing the order size.")
    print()
    
    # Select random cardholders
    selected_cardholders = grouped.sample(n=no_of_customers_to_target_rounded, random_state=42)

    # Calculate the sum of avg transaction value and cashback value for selected cardholders
    sum_avg_transaction = selected_cardholders['Total_Transaction_Value'].sum()
    sum_avg_cashback = selected_cardholders['Total_Cashback_Value'].sum()

    # Calculate profit from selected cardholders
    profit_from_selected = sum_avg_transaction - sum_avg_cashback

    # Display the selected cardholders and their sums
    print("Selected Cardholders:")
    print(tabulate(selected_cardholders, headers='keys', tablefmt='pretty', showindex=False))
    print()
    print(f"Sum of Selected Cardholders' Avg Transaction Value: {sum_avg_transaction}")
    print(f"Sum of Selected Cardholders' Avg Cashback Value: {sum_avg_cashback}")
    print(f"Profit from Selected Cardholders: {profit_from_selected}")

    # Additional conditional check
    if profit_from_selected >= revenue_target:
        print(f"Yes, we achieved the target successfully with randomly selected {no_of_customers_to_target_rounded} customers!")
    else:
        print(f"Sorry, we cannot achieve the target with randomly selected {no_of_customers_to_target_rounded} customers!")

    print("\n" + "="*120 + "\n")

# List of CSV files
cluster_names = ['Cluster 0', 'Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4']
files = [f'../Data/cluster_calculation/Full Dataset of {name}.csv' for name in cluster_names]

# Set the current sales and percentage increase
current_sales = 10000  # You can change this value as needed
percentage_increase = 20  # You can change this value as needed

# Iterate through each file and compute metrics
for file in files:
    print(f"Processing file: {file}")
    df = pd.read_csv(file)
    compute_metrics(df, current_sales, percentage_increase)
```
